```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./tcb.csv')
df['time'] = pd.to_datetime(df['time'], unit='ms')
df.set_index('time', inplace=True)
df['ssthresh_diff'] = df['ssthresh'].diff()
df['cwnd_diff'] = df['cwnd'].diff()
df = df.head(1100)
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['ssthresh'], label='ssthresh')
plt.plot(df.index, df['cwnd'], label='cwnd')

# Titel und Beschriftungen hinzufügen
plt.title('Zeitlicher Verlauf von sshtresh und cwnd')
plt.xlabel('Zeit')
plt.ylabel('Werte')
plt.legend()

# Zeige den Plot an
plt.show()
```
